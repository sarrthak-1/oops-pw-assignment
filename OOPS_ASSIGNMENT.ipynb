# **THEORY** **QUESTIONS**

1. What is Object-Oriented Programming (OOP)?


OOP is a programming style where we structure code using objects — each object represents real-world entities with their own data (attributes) and actions (methods). It helps make code reusable, organized, and easy to maintain.

2. What is a class in OOP?


A class is like a blueprint or template for creating objects. It defines how the objects will behave and what data they will hold.

3. What is an object in OOP?


An object is an instance of a class. It’s a real thing created from a class blueprint that can use the class’s attributes and methods.

4. What is the difference between abstraction and encapsulation?



Abstraction hides unnecessary details and only shows what’s needed.

Encapsulation keeps data safe by restricting direct access to it (using private variables and methods).

5. What are dunder methods in Python?


“Dunder” stands for “double underscore.” These are special built-in methods like __init__, __str__, __len__, etc., that let us define how objects behave with operators or built-in functions.

6. Explain the concept of inheritance in OOP.


Inheritance allows one class (child) to use the properties and methods of another class (parent). It helps in reusing code and building relationships between classes.0

7. What is polymorphism in OOP?


Polymorphism means “many forms.” It allows the same method name to behave differently depending on the object calling it — for example, different classes can have their own version of a method called speak().

8. How is encapsulation achieved in Python?


Encapsulation is done by making attributes private using underscores (like _balance or __balance) and accessing them only through getter and setter methods.

9. What is a constructor in Python?


A constructor is a special method named __init__() that runs automatically when a new object is created. It initializes the object’s attributes.

10. What are class and static methods in Python?

Class method (@classmethod) works with the class itself, not individual objects.

Static method (@staticmethod) doesn’t depend on either class or object and is used for general utility functions.

11. What is method overloading in Python?


Python doesn’t support true method overloading like Java, but we can achieve it using default parameters or *args to handle multiple types of arguments in one method.

12. What is method overriding in OOP?


When a child class defines a method with the same name as in its parent class, it replaces (or overrides) the parent’s version.

13. What is a property decorator in Python?


@property is used to make a method behave like an attribute — it lets us access values without using parentheses, often used for getters and setters.

14. Why is polymorphism important in OOP?


It helps make programs flexible and extendable — we can use the same interface for different types of objects without rewriting code.

15. What is an abstract class in Python?


An abstract class is a class that can’t be instantiated directly. It contains abstract methods (declared but not defined) and is created using the abc module.

16. What are the advantages of OOP?

Reusability of code

Easier to maintain

Better organization

Easier debugging and teamwork

17. What is the difference between a class variable and an instance variable?



Class variable: shared by all objects of the class.

Instance variable: unique for each object.

18. What is multiple inheritance in Python?


When a class inherits from more than one parent class, it’s called multiple inheritance.

19. Explain the purpose of the __str__ and __repr__ methods in Python.

__str__() → used for readable, user-friendly string representation.

__repr__() → used for official or developer-friendly string output.

20. What is the significance of the super() function in Python?


super() lets us call methods or constructors of the parent class, so we can extend or reuse the parent’s functionality.

21. What is the significance of the __del__ method in Python?


It’s the destructor method that runs automatically when an object is deleted, usually used to clean up resources.

22. What is the difference between @staticmethod and @classmethod in Python?

@staticmethod → doesn’t take self or cls; used for general utility.

@classmethod → takes cls as the first parameter and can access class-level data.

23. How does polymorphism work in Python with inheritance?


When a child class overrides a method from the parent, and we call that method using a parent reference, Python runs the child version — that’s runtime polymorphism.

24. What is method chaining in Python OOP?


It means calling multiple methods in one line — by returning self from each method

25. What is the purpose of the __call__ method in Python?


It lets an object behave like a function — when you use parentheses after an object, Python runs the __call__ method inside it.

```python
# 1. Create a parent class Animal with a method speak() and a child class Dog that overrides it to print "Bark!".
class Animal:
    def speak(self):
        print("This animal makes a sound.")

class Dog(Animal):
    def speak(self):
        print("Bark!")

dog = Dog()
dog.speak()
```

```python
# 2. Create an abstract class Shape with a method area(), and derive Circle and Rectangle classes implementing it.
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

circle = Circle(5)
rectangle = Rectangle(4, 6)
print("Circle area:", circle.area())
print("Rectangle area:", rectangle.area())
```

```python
# 3. Implement multi-level inheritance with Vehicle, Car, and ElectricCar classes adding a battery attribute.
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery):
        super().__init__(vehicle_type, brand)
        self.battery = battery
    def info(self):
        print(f"{self.brand} is an {self.vehicle_type} with a {self.battery} battery.")

tesla = ElectricCar("car", "Tesla", "100kWh")
tesla.info()
```

```python
# 4. Demonstrate polymorphism using a base class Bird and derived classes Sparrow and Penguin overriding fly().
class Bird:
    def fly(self):
        print("Birds can usually fly.")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow is flying high!")

class Penguin(Bird):
    def fly(self):
        print("Penguins can't fly, they swim!")

for bird in [Sparrow(), Penguin()]:
    bird.fly()
```

```python
# 5. Demonstrate encapsulation using a BankAccount class with private balance and deposit/withdraw methods.
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited ₹{amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ₹{amount}")
        else:
            print("Insufficient balance!")

    def get_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(300)
print("Current balance:", acc.get_balance())
```

```python
# 6. Demonstrate runtime polymorphism using a base class Instrument and derived classes Guitar and Piano.
class Instrument:
    def play(self):
        print("Playing some instrument sound...")

class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar!")

class Piano(Instrument):
    def play(self):
        print("Playing the piano keys!")

for instrument in [Guitar(), Piano()]:
    instrument.play()
```

```python
# 7. Create a MathOperations class with a class method add_numbers() and a static method subtract_numbers().
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print("Addition:", MathOperations.add_numbers(10, 5))
print("Subtraction:", MathOperations.subtract_numbers(10, 5))
```

```python
# 8. Implement a Person class with a class method to count the total number of persons created.
class Person:
    count = 0

    def __init__(self, name):
        self.name = name
        Person.count += 1

    @classmethod
    def total_persons(cls):
        return cls.count

p1 = Person("Sarthak")
p2 = Person("Ananya")
print("Total persons created:", Person.total_persons())
```

```python
# 9. Create a Fraction class with numerator and denominator and override __str__ to display it as 'num/den'.
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

frac = Fraction(3, 4)
print(frac)
```

```python
# 10. Demonstrate operator overloading by creating a Vector class and overriding the __add__ method.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1 + v2)
```

```python
# 11. Create a Person class with name and age attributes and a greet() method that prints an introduction.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

p = Person("Sarthak", 23)
p.greet()
```

```python
# 12. Create a Student class with name and grades attributes and a method average_grade() to compute average.
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

stu = Student("Sarthak", [85, 90, 95])
print(f"{stu.name}'s average grade is {stu.average_grade()}")
```

```python
# 13. Create a Rectangle class with set_dimensions() and area() methods to calculate the area.
class Rectangle:
    def set_dimensions(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

rect = Rectangle()
rect.set_dimensions(5, 10)
print("Area of rectangle:", rect.area())
```

```python
# 14. Create an Employee class with calculate_salary(), and a Manager class adding a bonus to the salary.
class Employee:
    def __init__(self, hours, rate):
        self.hours = hours
        self.rate = rate

    def calculate_salary(self):
        return self.hours * self.rate

class Manager(Employee):
    def __init__(self, hours, rate, bonus):
        super().__init__(hours, rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

mgr = Manager(40, 200, 5000)
print("Manager total salary:", mgr.calculate_salary())
```
